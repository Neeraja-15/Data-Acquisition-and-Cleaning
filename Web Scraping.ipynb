```python
from bs4 import BeautifulSoup

# Creating an HTML document
html_doc = """<html><head><title>The Dormouse's story</title></head>
<body>
<p class="title"><b>The Dormouse's story</b></p>
<p class="story">Once upon a time there were three little sisters; and their names were
<a href="http://example.com/elsie" class="sister" id="link1">Elsie</a>,
<a href="http://example.com/lacie" class="sister" id="link2">Lacie</a>
and
<a href="http://example.com/tillie" class="sister" id="link3">Tillie</a>;
and they lived at the bottom of a well.</p>
<p class="story">...</p>
</body>
</html>"""

# Parsing the HTML
soup = BeautifulSoup(html_doc, 'html.parser')

# Prettify the HTML structure
print(soup.prettify())

# Accessing different parts of the parsed HTML
title_tag = soup.title  # <title>The Dormouse's story</title>
print(title_tag)
print(title_tag.name)  # 'title'
print(title_tag.string)  # "The Dormouse's story"
print(title_tag.parent.name)  # 'head'

# Accessing the first anchor tag
first_anchor = soup.a
print(first_anchor)  # <a class="sister" href="http://example.com/elsie" id="link1">Elsie</a>

# Finding all anchor tags
all_anchors = soup.find_all('a')
print(all_anchors)

# Finding an element by ID
link3 = soup.find(id='link3')
print(link3)  # <a class="sister" href="http://example.com/tillie" id="link3">Tillie</a>

# Extracting href attributes from all anchor tags
for iterator in soup.find_all('a'):
    print(iterator.get('href'))

# Extracting id attributes from all anchor tags
for iterator in soup.find_all('a'):
    print(iterator.get('id'))

# Extracting text content of the entire document
print(soup.get_text())

```
