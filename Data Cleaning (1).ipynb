Imputation (Filling missing values)

````python
df
````

````python
# Calculate the mean of the column 'Quantity'
mean_quantity= df['Quantity'].mean()
mean_quantity
1.6111111111111112
# Fill missing values in 'Quantity' with the mean of the column
df['Quantity'] = df['Quantity'].fillna(mean_quantity)
df
````

````python
# Calculate the median of the column 'Price'
median_price = df['Price'].median()
median_price
# Fill missing values in 'Price' with the median
df['Price'] = df['Price'].fillna(median_price)
df
````

````python
# Identify the mode within a column (most frequent value)
mode_product = df['Product'].mode()[0]
mode_product
# Fill missing values in 'Product' with the mode
df['Product'] = df['Product'].fillna(mode_product)
df
````
